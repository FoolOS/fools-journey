```python
"Wizard Picker 9000"

#The list for main HP Chars
def good():
    return ['Harry', 'Ron', 'Hermione']

def odds():
    for number in range(10):
        if number % 2 != 0:
            yield number
print("The good characters:", good())

count = 0
for odd in odds():
    count += 1
    if count == 3:
        print("Third value:", odd)
        break
```
