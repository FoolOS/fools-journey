```python
class Solution:

    def binarysearch(self, arr, k):
        left, right = 0, len(arr) - 1
        result = -1

        while left <= right:
            mid = left + (right - left) // 2

            if arr[mid] == k:
                result = mid
                right = mid - 1
            elif arr[mid] < k:
                left = mid + 1
            else:
                right = mid - 1

        return result

n = int(input("Enter number of elements in the array: "))  
arr = list(map(int, input("Enter sorted array, making sure to leave spaces between elements and no commas: ").split()))
k = int(input("Enter target number in the array: "))

sol = Solution()
print("Index of first occurence: ", sol.binarysearch(arr, k))
```
