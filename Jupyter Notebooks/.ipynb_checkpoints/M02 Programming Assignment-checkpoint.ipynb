```python
"Guessing Game"

secret = 4
guess = int(input("Guess the secret number! Pick a number from 1 to 10: "))

#Logic for picking too low of a number outside of range.
if guess < 1:
    print("That's below 1! Pick a different number from 1 to 10!!!")

elif guess < secret:
    print("Whoops! Too low! Try again!")
elif guess == secret:
    print("Exactly right!")

#Logic for picking a number above ten.
elif guess > 10:
    print("That's bigger than 10! Pick a different number from 1 to 10!!!")

elif guess > secret:
    print("Dang, too high! Try again!")
```

```python
"Small and Green Fruits"

small = True
green = True

#Cherry
if small and not green:
    print("Cherry matches.")
else:
    print("Cherries aren't green.")
#Pea
if small and green:
    print("Peas are small and green.")
else:
    print("Peas don't match.")
#Watermelon
if not small and green:
    print("Watermelon matches.")
else:
    print("Watermelons aren't usually small.")
#Pumpkin
if not small and not green:
    print("Pumpkin matches.")
else:
    print("Pumpkins aren't usually small and green.")
```

```python
"Countdown"

for x in list( range(3, -1, -1)):
    print(x)
```

```python
"Incremental Guessing Game"

guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("Too low!")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Whoopsie doopsie!")
        break
    number += 1
```

```python
"For-Loop Guessing Game"

guess_me = 5

#Change the 1 in range for different starting number.
for number in range(1, 10):
    if number < guess_me:
        print("Too low!")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Whoops...")
        break
```
